Pick 7 categories from the KickStarterCombo Dataset
Build a predictive model for success for each individual category. Determine which one is most useful.

```python
import zipfile         ## open zipfile
path_to_zip_file = "Kickstartercombo.zip"
zip_ref = zipfile.ZipFile(path_to_zip_file,'r')
zip_ref.extractall("")
zip_ref.close()
```

```python
import pandas as pd    ## open zipfile to read csv file
Location = "Kickstartercombo.zip"
df=pd.read_csv(Location)

df.head()
```

```python
def score_to_numeric(x):          ## give 1 to successful projects, otherwise, give 0
    if x == 'successful':
        return 1
    if x != 'successful':
        return 0
```

```python
df['state_success'] = df['state'].apply(score_to_numeric)  ## add a new column to show 'successful state' with numeric
df.head()
```

```python
df.groupby('main_category')['state_success'].mean()    ## calculate mean per group(main_category)
```

```python
catdf = pd.DataFrame(df.groupby('main_category')['state_success'].mean())    
catdf       ## get a new dataframe to show mean per each main_catergory before we get the rankings of mean               
```

```python
catdf = catdf.sort_values('state_success', ascending=False)
catdf.head()   ## sorting for getting rankings
```

```python
## calculate mean of success per sub_category(column 'category')
df.groupby('category')['state_success'].mean()    
```

```python
## new dataframe to get rankings per group
subcatdf = pd.DataFrame(df.groupby('category')['state_success'].mean())
subcatdf.head()
```

```python
## sorting for getting 7 top ranked category
subcatdf = subcatdf.sort_values('state_success', ascending =False)
subcatdf.head(7)
```

```python
 ####removing time from date&time variable for simple calcucating

from datetime import date   
df['deadline'] = pd.to_datetime(df['deadline']).dt.date
df['launched'] = pd.to_datetime(df['launched']).dt.date
df.head()
```

```python
## create 'duration' variable to see if the duration affects the success.

from datetime import date
a = df['launched']
b = df['deadline']
df['duration'] = (b - a)
df.head()
```

```python
## convert 'duration' to numbers to run  regressions

import numpy as np
df['date_delta'] = df['duration'] / np.timedelta64(1,'D')
df.head()
```

```python
## run an OLS regression with whole data
import statsmodels.formula.api as sm
result = sm.ols(formula='state_success ~ goal + pledged+ backers +date_delta', data=df).fit()
result.summary()
```
